```python
from pathlib import Path
import torch
import numpy as np; np.random.seed()
import matplotlib.pyplot as plt
import seaborn as sns

import os
import math

os.environ['KMP_DUPLICATE_LIB_OK']='True'
%matplotlib inline 
```

```python
data_path = Path("./data/imdb_image")
labels_path = Path("./data")
data = torch.load(data_path /"imdb_images-6to12.pt")
labels = torch.load(labels_path /"labels2D.pt")
```

```python
data = np.array(data)
```

```python
print(data[0].shape)
```

```python
def z_normalize(input_tensor): # layerNorm from paper
    input_tensor = torch.from_numpy(input_tensor)
    mean = input_tensor.mean()
    std = input_tensor.std()
    up = torch.sub(input_tensor, mean)
    down = torch.add(std**2, 0.001**2)
    return np.array(torch.div(up, math.sqrt(down)))
```

```python
"""
IMDB-Image[0:5]
"""
fig, ax = plt.subplots(1, 5, figsize=(15, 7))
for i in range(0,1):
    ax[i].imshow(data[i])
    ax[i].title.set_text("Label : {}".format(labels[i]))
    ax[i+1].imshow(data[i+1])
    ax[i+1].title.set_text("Label : {}".format(labels[i+1]))
    ax[i+2].imshow(data[i+2])
    ax[i+2].title.set_text("Label : {}".format(labels[i+2]))
    ax[i+3].imshow(data[i+3])
    ax[i+3].title.set_text("Label : {}".format(labels[i+3]))
    ax[i+4].imshow(data[i+4])
    ax[i+4].title.set_text("Label : {}".format(labels[i+4]))
    
plt.tight_layout()
```

```python
"""
Applying Z-Normalization to IMDB-image[0:5]
"""

fig, ax = plt.subplots(1, 5, figsize=(15, 7))
for i in range(0,1):
    ax[i].imshow(z_normalize(data[i]))
    ax[i].title.set_text("Label : {}".format(labels[i]))
    ax[i+1].imshow(z_normalize(data[i+1]))
    ax[i+1].title.set_text("Label : {}".format(labels[i+1]))
    ax[i+2].imshow(z_normalize(data[i+2]))
    ax[i+2].title.set_text("Label : {}".format(labels[i+2]))
    ax[i+3].imshow(z_normalize(data[i+3]))
    ax[i+3].title.set_text("Label : {}".format(labels[i+3]))
    ax[i+4].imshow(z_normalize(data[i+4]))
    ax[i+4].title.set_text("Label : {}".format(labels[i+4]))
    
plt.tight_layout()
```
